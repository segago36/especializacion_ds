```python
# Aplicando el principio de segregación de interfaces
class Impresora:
    def imprimir(self, documento):
        pass


class Escaner:
    def escanear(self):
        pass


class Fotocopiadora:
    def fotocopiar(self, documento):
        pass


class ImpresoraMultifuncional(Impresora, Escaner, Fotocopiadora):
    pass


class Oficina:
    def __init__(self, impresora):
        self.impresora = impresora

    def imprimir_documento(self, documento):
        self.impresora.imprimir(documento)

    def escanear_documento(self):
        if isinstance(self.impresora, Escaner):
            self.impresora.escanear()
        else:
            print("La impresora no es compatible con la función de escaneo.")

    def fotocopiar_documento(self, documento):
        if isinstance(self.impresora, Fotocopiadora):
            self.impresora.fotocopiar(documento)
        else:
            print("La impresora no es compatible con la función de fotocopiado.")
```
